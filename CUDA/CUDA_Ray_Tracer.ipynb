## Performance Benchmark (Optional)

Compare CUDA performance with different configurations

```python
# Run multiple times to see consistent performance
import time

print("Running benchmark (5 iterations)...")
times = []

for i in range(5):
    start = time.time()
    !./raytrace_cuda > /dev/null 2>&1
    elapsed = time.time() - start
    times.append(elapsed)
    print(f"Run {i+1}: {elapsed:.3f}s")

print(f"\nAverage time: {sum(times)/len(times):.3f}s")
print(f"Min time: {min(times):.3f}s")
print(f"Max time: {max(times):.3f}s")
```
